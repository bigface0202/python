```python
import pandas as pd
import numpy as np
from IPython.core.pylabtools import figsize
from matplotlib import pyplot as plt
import collections
%matplotlib inline
from scipy.optimize import curve_fit
import seaborn as sns
import pymc as pm
```

```python
############################################
#######全被験者毎でまとめたデータの作図#####
############################################

Subject_list = [1,2,3]
Stimuli_list = [30,60,90,-30,-60,-90]
def sigmoid(x, x0, k):
     y = 1 / (1 + np.exp(-k*(x-x0)))
     return y
for i in Subject_list:
    for j in Stimuli_list:
        Exp_DF = pd.read_csv("sub" + str(i) + "/" + str(i) + "_" + str(j) + ".csv")#csvデータの読み込み
        Exp_DF_group = Exp_DF.groupby("Stimulus")[['Answer']].count()#まずは刺激の回数をカウント
        Exp_DF_group["Count of heavy"] = Exp_DF.groupby("Stimulus")[['Answer']].sum()/abs(j)#刺激に対して実験値の方が重く感じた回数をカウント
        Exp_DF_group["Fraction of heavy"] = Exp_DF_group["Count of heavy"]/Exp_DF_group["Answer"]#刺激の回数に対して重く感じた割合
    
        #作図
        plt.figure()
        plt.scatter(Exp_DF_group.index, Exp_DF_group["Fraction of heavy"])
    
        ##カーブフィッティング
        x0_initial = 10.2
        k_initial = 0.2
        ###Scipyのカーブフィッティング関数に値を渡す．
        try:
            popt, pcov = curve_fit(sigmoid, Exp_DF_group.index, Exp_DF_group["Fraction of heavy"],[x0_initial,k_initial])
    
        except RuntimeError:
            print("Error - curve_fit failed")
        xs_fit = np.linspace(110,290 ,100)
        print("Subject" + str(i) + "_" + str(j) + "deg.:",popt)#シグモイドのパラメータを表示
        ys_fit = sigmoid(xs_fit, *popt)
        plt.plot(xs_fit,ys_fit)#フィッティングしたグラフを表示

        plt.xlabel("Stimulus of Torque")
        plt.xlim(100,300)
        plt.ylim(-0.1,1.2)
        plt.title("Sub." + str(i) + "_"+str(j) + "degree")
```

```python
############################################
##全被験者を角度毎でまとめたデータの作成####
############################################
def sigmoid(x, x0, k):
     y = 1 / (1 + np.exp(-k*(x-x0)))
     return y

Angle_list=[]
for i in Stimuli_list:
    Fraction_list = []
    for j in Subject_list:
        Exp_DF = pd.read_csv("sub" + str(j) + "/" + str(j) + "_" + str(i) + ".csv")#csvデータの読み込み
        Exp_DF_group = Exp_DF.groupby("Stimulus")[['Answer']].count()#まずは刺激の回数をカウント
        Exp_DF_group["Count of heavy"] = Exp_DF.groupby("Stimulus")[['Answer']].sum()/abs(i)#刺激に対して実験値の方が重く感じた回数をカウント
        Exp_DF_group["Fraction of heavy"] = Exp_DF_group["Count of heavy"]/Exp_DF_group["Answer"]#刺激の回数に対して重く感じた割合
        Fraction_list.append(Exp_DF_group)
    Subject_concat = sum(Fraction_list)
    Subject_concat["Fraction of heavy"] = Subject_concat["Count of heavy"] / Subject_concat["Answer"]
    Angle_list.append(Subject_concat)

Result_DF = pd.DataFrame({"30":Angle_list[0]["Fraction of heavy"],
                         "60":Angle_list[1]["Fraction of heavy"],
                         "90":Angle_list[2]["Fraction of heavy"],
                         "-30":Angle_list[3]["Fraction of heavy"],
                         "-60":Angle_list[4]["Fraction of heavy"],
                         "-90":Angle_list[5]["Fraction of heavy"]},
                        columns=['30', '60', '90','-30','-60','-90'])

######################################
##全被験者を角度毎でまとめたグラフ####
######################################
def sigmoid(x, x0, k):
     y = 1 / (1 + np.exp(-k*(x-x0)))
     return y
#角度でまとめたデータをプロット
Stimuli_list = [30,60,90,-30,-60,-90]
for i in Stimuli_list:
    plt.figure()
    plt.scatter(Result_DF.index,Result_DF[str(i)])
    ##カーブフィッティング
    x0_initial = 10.2
    k_initial = 0.2
    ###Scipyのカーブフィッティング関数に値を渡す．
    try:
        popt, pcov = curve_fit(sigmoid, Result_DF.index, Result_DF[str(i)],[x0_initial,k_initial])
    
    except RuntimeError:
        print("Error - curve_fit failed")
    xs_fit = np.linspace(110,290 ,100)
    print(str(i) + "deg.:",popt)#シグモイドのパラメータを表示
    ys_fit = sigmoid(xs_fit, *popt)
    plt.plot(xs_fit,ys_fit)#フィッティングしたグラフを表示
    
    plt.plot([100,300],[0.5,0.5], linestyle = 'dashed')#PSE？の位置
    plt.plot([popt[0],popt[0]],[-0.1,1.2], linestyle = 'dashed')#PSE?の位置

    plt.xlabel("Stimulus of Torque",fontsize = 17)
    plt.ylabel("Fraction larger steering force",fontsize = 17)
    plt.xlim(100,300)
    plt.ylim(-0.1,1.2)
    plt.title(str(i) + "degree",fontsize = 17)
```

```python
############################################
##全被験者を角度毎でまとめたヒストグラム####
###########################################

Row_data_list=[]
for i in Stimuli_list:
    Fraction_list = []
    for j in Subject_list:
        Exp_DF = pd.read_csv("sub" + str(j) + "/" + str(j) + "_" + str(i) + ".csv")#csvデータの読み込み
        Exp_DF["Answer"] = Exp_DF["Answer"]/abs(i)
        Fraction_list.append(Exp_DF)
    Tmp_DF = pd.concat(Fraction_list).dropna(axis = 1)
    Row_data_list.append(Tmp_DF)#Stimuli_listの順番にリストへ格納される0:30,1:60,2:90,3:-30,4:-60,5:-90 deg

for i,j in enumerate(Stimuli_list):
    plt.figure()
    plt.hist(Row_data_list[i]["Stimulus"],normed=True,color="red",alpha=0.5)
    plt.hist(Row_data_list[i][Row_data_list[i].Answer != 0]["Stimulus"], normed=True, alpha=0.5)#重いと答えた重さ
    plt.title(str(j)+'degree')
    plt.xlabel("Torque stimuli")
    plt.ylabel("Frequency")
```
